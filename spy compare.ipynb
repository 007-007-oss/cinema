```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "name": ["bond","bourne"],
    "agency": ["mi6","cia"],
    "salary": [60000 , 84000],
    "kill count" : [597,21],
    "mission": [26 , 4]
})
print(df.head())

#for finding average kill count,divide kill count by mission by each
average_kill_count = df["kill count"] / df["mission"]

# for finding average kill count,divide kill count by mission
display(average_kill_count)

# find their average salary per mission
average_salary_per_mission = df["salary"] / df["mission"]
display(average_salary_per_mission)

# find their average salary acc to kill count
average_salary_acc_to_kill_count = df["salary"]/df["kill count"]
display(average_salary_acc_to_kill_count)


```
